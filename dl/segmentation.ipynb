```python
# Import required libraries
import torch

# Load YOLOv5 model (YOLOv5s is a small, faster model; you can also use YOLOv5m, YOLOv5l, or YOLOv5x)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
print("YOLOv5 model loaded successfully.")
```

```python
# Perform object detection on the specified image
results = model(img_path)
print("Object detection completed.")
```

```python
# Display results
results.show()  # Shows the image with detected objects using matplotlib
```

```python
# Extract details from detection results
detections = results.pandas().xyxy[0]  # Get detections as a DataFrame

# Display the detection DataFrame
detections.head()  # This will show the first few rows of detected objects
```

```python
# Filter detections with a confidence score above 0.5
filtered_detections = detections[detections['confidence'] > 0.5]
filtered_detections  # Display filtered results
```

```python
#Depth calculations
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms

# Check if the MiDaS model is available locally and load it
try:
    midas = torch.hub.load("intel-isl/MiDaS", "DPT_Large")
except:
    print("MiDaS model not found locally. Downloading and saving it for future use.")
    midas = torch.hub.load("intel-isl/MiDaS", "DPT_Large", force_reload=True)

midas.eval()

# Load the transformations for MiDaS
try:
    transform = torch.hub.load("intel-isl/MiDaS", "transforms").dpt_transform
except:
    print("Transformations not found locally. Downloading.")
    transform = torch.hub.load("intel-isl/MiDaS", "transforms", force_reload=True).dpt_transform

# Load the image using the previously saved img_path variable
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply MiDaS transformations to the image
input_batch = transform(img)

# Run MiDaS model to get the depth map
with torch.no_grad():
    depth_prediction = midas(input_batch)

# Process the depth map
depth_map = depth_prediction.squeeze().cpu().numpy()
depth_map = cv2.resize(depth_map, (img.shape[1], img.shape[0]))

# Check the shape and range of the depth map for debugging
print("Depth map shape:", depth_map.shape)
print("Depth map min value:", depth_map.min())
print("Depth map max value:", depth_map.max())

# Normalize depth map values to range [0, 1]
depth_map_normalized = (depth_map - depth_map.min()) / (depth_map.max() - depth_map.min())

# Display the normalized depth map
plt.imshow(depth_map_normalized, cmap="inferno")
plt.colorbar()
plt.title("Normalized Depth Map")
plt.show()
```

```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt

# Initialize MediaPipe Pose model for human pose estimation
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True)

# Load the image with a person
img_path = "path/to/your/human_image.jpg"
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for mediapipe

# Process the image to detect human pose landmarks
results = pose.process(image_rgb)

# Draw landmarks if detected
if results.pose_landmarks:
    print("Human pose detected.")
    annotated_image = image_rgb.copy()
    mp_drawing = mp.solutions.drawing_utils
    mp_drawing.draw_landmarks(annotated_image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Display the image with the skeleton
    plt.imshow(annotated_image)
    plt.axis("off")
    plt.title("Human Skeletal Pose Estimation")
    plt.show()
else:
    print("No human pose detected in the image.")
```
